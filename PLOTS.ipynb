```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
age = pd.read_csv(r'C:\Users\gntai\Desktop\prevent\PREVENT_datasets_26082022\ecrf_assessments\ecrf_demographics.csv')
age
```

```python
apoe = pd.read_csv(r'C:\Users\gntai\Desktop\prevent\PREVENT_datasets_26082022\apoe_version2.csv')
apoe
```

```python
apoe = apoe.dropna(subset=['apoe4'])
apoe
```

```python
combined_df = pd.merge(age, apoe, on='subjectID')
combined_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming combined_df is your merged dataframe

# Filter combined_df to include only rows with ages between 40 and 60
filtered_df = combined_df[(combined_df['age'] >= 40) & (combined_df['age'] <= 60)]

# Start plotting
fig, axs = plt.subplots(1, 3, figsize=(18, 5), gridspec_kw={'wspace':0.3})

# Customize font sizes
label_size = 14
title_size = 16
tick_size = 13  # New variable for tick sizes

# Plot A: Sex by site in reversed grayscale
sns.countplot(x='centre', hue='sex', data=combined_df, ax=axs[0], palette='Greys_r')
axs[0].set_title('Sex by site', fontsize=title_size)
axs[0].set_xlabel('Site', fontsize=label_size)
axs[0].set_ylabel('Count', fontsize=label_size)
axs[0].legend(title='Sex')
axs[0].text(-0.1, 1.1, 'A', transform=axs[0].transAxes, size=20, weight='bold')
axs[0].tick_params(labelsize=tick_size)  # Set the tick size

# Plot B: Density of age by site, limited to ages between 40 and 60
sns.kdeplot(data=filtered_df, x='age', hue='centre', fill=True, common_norm=False, ax=axs[1])
axs[1].set_title('Density of age by site', fontsize=title_size)
axs[1].set_xlabel('Age', fontsize=label_size)
axs[1].set_ylabel('Density', fontsize=label_size)
axs[1].set_xlim(40, 60)  # Setting the x-axis limit
axs[1].text(-0.1, 1.1, 'B', transform=axs[1].transAxes, size=20, weight='bold')
axs[1].tick_params(labelsize=tick_size)  # Set the tick size

# Plot C: APOE by site in reversed grayscale
sns.countplot(x='centre', hue='apoe_status', data=combined_df, ax=axs[2], palette='Greys_r')
axs[2].set_title('APOE by site', fontsize=title_size)
axs[2].set_xlabel('Site', fontsize=label_size)
axs[2].set_ylabel('Count', fontsize=label_size)
axs[2].legend(title='APOE Status')
axs[2].text(-0.1, 1.1, 'C', transform=axs[2].transAxes, size=20, weight='bold')
axs[2].tick_params(labelsize=tick_size)  # Set the tick size

# Adjust top padding of the subplots
plt.subplots_adjust(top=0.8)

# Save the entire figure
fig.savefig('combined_plots5.png')

plt.show()
```
